## Modelling

We already have df_x and y data with us, so no need to split the data, we can proceed with train and test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df_x,y,test_size=.22,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr
```

```python
lr.fit(x_train,y_train)
lr.score(x_train,y_train)
pred=lr.predict(x_test)
pred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
print('Mean squared error: ',mean_squared_error(y_test,pred))
print('Mean absolute error: ',mean_squared_error(y_test,pred))
```

```python
from sklearn.metrics import r2_score
print('R2 socre: ',r2_score(y_test,pred))
```

## Regularisation

## Lasso and ridge

```python
from sklearn.linear_model import Lasso,Ridge
```

```python
# alpha values could be .001,.01,.1,1,10 etc., higher alpha values reduces all the coefficient to 0 and impact the output
# Default value of alpha=1.0
ls=Lasso(alpha=0.0001)
ls.fit(x_train,y_train)
ls.score(x_train,y_train)
```

```python
ls.coef_
```

```python
# try to minimize the coef value
rd=Ridge(alpha=0.0001)
rd.fit(x_train,y_train)
rd.score(x_train,y_train)
```

## Elassticnet

```python
from sklearn.linear_model import ElasticNet
ent=ElasticNet(alpha=0.0001)
ent.fit(x_train,y_train)
ent.score(x_train,y_train)
```

We got to see that the regularisaction is also giving almost the same acccuracy hence trying one more regressor to rechek.

```python
from sklearn.svm import SVR
```

```python
svr= SVR(kernel='linear')
svr.fit(x_train,y_train)
svr.score(x_train,y_train)
#99.5
pred=svr.predict(x_test)
pred
```

```python
Ker=['linear','poly','rbf']
for i in Ker:
    sv=SVR(kernel=i)
    sv.fit(x_train,y_train)
    print("score",sv.score(x_train,y_train))
```

So from the above details we see here that the accuracy is best in linear regression model and hence finalising the model.
